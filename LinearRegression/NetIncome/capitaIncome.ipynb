```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
%matplotlib inline
plt.ylabel('price')
plt.xlabel('year')
plt.scatter(df['year'], df['per capita income (US$)'], color='blue')
```

```python
income = df['per capita income (US$)']
income.head(5)
```

```python

model = linear_model.LinearRegression()
model.fit(df[['year']], df['per capita income (US$)'])
predictions = model.predict(income) #Making new column for predicted prices
df['new_prices'] = predictions
df.to_csv('predictions.csv', index=False) #exporting to a new csv file.
```

```python
%matplotlib inline
plt.ylabel('price')
plt.xlabel('year')
plt.scatter(df['year'], df['per capita income (US$)'], color='blue')
plt.plot(df['year'], model.predict(df[['year']]), color='red')
```
