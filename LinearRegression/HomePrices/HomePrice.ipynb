```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv') #Loading the dataset
df #Reading the dataset
```

```python
%matplotlib inline  
plt.pyplot.xlabel('area(sqr ft)') #Plotting the data
plt.pyplot.ylabel('price(US$)')
plt.pyplot.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression() #Creating model
reg.fit(df[['area']],df.price) #Loading data into the model, It should be 2d frame
```

```python
reg.predict([[4000]])
```

```python
%matplotlib inline  
plt.pyplot.xlabel('area(sqr ft)') #Plotting the data
plt.pyplot.ylabel('price(US$)')
plt.pyplot.scatter(df.area, df.price, color='red', marker='+')
plt.pyplot.plot(df.area, reg.predict(df[['area']]), color='green') #Adding line of best fit
```

```python
d = pd.read_csv("areas.csv") # reading just prices
d.head(5)
```

```python
predictions = reg.predict(d) #Loading all prices into our model
d['prices'] = predictions #Making new column for predicted prices
d.to_csv('predictions.csv', index=False) #exporting to a new csv file.
```
